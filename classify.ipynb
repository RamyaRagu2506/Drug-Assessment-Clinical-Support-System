RECOMMENDATION FUNCTION 


```python
def userful_count(data,feature):
 max_value = data[feature].max()
 min_value = data[feature].min()
 data[feature] = (data[feature] - min_value) / (max_value - min_value)
 return data 
```

```python
def recom_sys (condition,label):
    data = userful_count(review_data,'usefulCount') 
    data['rec_score'] = (data[label]) * data['usefulCount']
    score = data[data['condition']==condition].sort_values(by=['rec_score'],ascending=False)
    return score['drugName'].unique()

```
