```python
import pandas as pd 
import numpy as np 
```

```python
processed_data = pd.read_csv("data/processed_data/drugsComTrain_feature_engineering.csv")
```

```python
processed_data.head()
```

```python
from textblob import TextBlob, Word, Blobber
from nltk.stem import PorterStemmer
st = PorterStemmer()
```

```python
rows_with_floats = processed_data[processed_data['stopwords_remove_review'].apply(lambda x: isinstance(x, float))]
```

```python
rows_with_floats
```

```python
# remove null values 
processed_data_new = processed_data[processed_data['stopwords_remove_review'].notna()]
```

```python
processed_data_new['stemmed_review'] = processed_data_new['stopwords_remove_review'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
processed_data_new.head()
```

```python
processed_data_new['lemmatized_review'] = processed_data_new['stopwords_remove_review'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
processed_data_new.head()
```

```python
processed_data_new['review_len'] = processed_data_new['stopwords_remove_review'].astype(str).apply(len)
processed_data_new['stemmed_review_len'] = processed_data_new['stemmed_review'].astype(str).apply(len)
processed_data_new['lem_review_len'] = processed_data_new['lemmatized_review'].astype(str).apply(len)
```

```python
processed_data_new.head()
```

```python
processed_data_new['word_count'] =processed_data_new['stopwords_remove_review'].apply(lambda x: len(str(x).split()))
```

```python
processed_data_new.head()
```

TO IDENTIFY SENTIMENT POLARITY 

VADER: 

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
def vader_sentiment(text):
    return sid.polarity_scores(text).get('compound')
```

```python
processed_data_new['sentiment_vader'] = processed_data_new['stopwords_remove_review'].apply(vader_sentiment)
```

```python
processed_data_new.head()
```

```python
def vader_classify(compound_score):
    if compound_score >= 0.05:
        sentiment= 'Positive'
    elif compound_score <= -0.05:
        sentiment= 'Negative'
    else:
        sentiment='Neutral'
    return sentiment
```

```python
processed_data_new['sentiment_vader_label'] = processed_data_new['sentiment_vader'].apply(vader_classify)
```

```python
processed_data_new.head()
```

```python
from textblob import TextBlob
```

```python

def Textblob_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
processed_data_new['sentiment_textblob'] = processed_data_new['stopwords_remove_review'].apply(Textblob_sentiment)
```

```python
processed_data_new.head()
```

```python
def textblob_classify(polarity):
    if polarity > 0:
        sentiment = 'positive'
    elif polarity == 0:
        sentiment = 'neutral'
    else:
        sentiment = 'negative'
    return sentiment
```

```python
processed_data_new['sentiment_textblob_label'] = processed_data_new['sentiment_textblob'].apply(textblob_classify)
```

```python
processed_data_new.head()
```

```python
processed_data_new = processed_data_new.drop(['Unnamed: 0'],axis=1)
```
