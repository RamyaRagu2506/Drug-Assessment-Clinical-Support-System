DATA-PREPROCESSING FOR REVIEW COLUMN (TEXT FORMAT)

```python
import spacy 
import nltk # natural language package similar to spacy 
import string
import re # regular expression
from nltk.corpus import stopwords
from tqdm import tqdm
import pandas as pd
import numpy as np 

```

```python
train_data_cleaned = pd.read_csv("data/processed_data/drugsComTrain_cleaned.csv")
```

```python
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase


```

```python
def preprocess_text(text_data):

    text_data = decontracted(text_data)
    
    text_data = text_data.replace('\n',' ')
    text_data = text_data.replace('\r',' ')
    text_data = text_data.replace('\t',' ')
    text_data = text_data.replace('-',' ')
    text_data = text_data.replace("/",' ')
    text_data = text_data.replace(">",' ')
    text_data = text_data.replace('"',' ')
    text_data = text_data.replace('?',' ')
    return text_data
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
  
# loading stop words from nltk library

stop_words = set(stopwords.words('english'))
stemmer = SnowballStemmer('english')

#removing 'no' from the stop words list as there is an importance of 'side effects' and 'no side effects' in review
stop_words.remove('no')
```

```python
#function for text lowering 
def text_lowering(text):
    return text.lower()
```

```python
train_data_cleaned['lowered_review'] = train_data_cleaned['review'].apply(text_lowering)
```

```python
train_data_cleaned.head()
```

```python
def punct_remove(text):
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator)
    return cleaned_text
```

```python
train_data_cleaned['remove_punctuation_review']=train_data_cleaned['lowered_review'].apply(punct_remove)
```

```python
train_data_cleaned.head()
```

```python
def remove_stopwords(review):
    '''This functional block preprocess the text data by removing digits, extra spaces, stop words 
    and converting words to lower case and stemming words'''
    
    if type(review) is not int:
        string = ""
        review = preprocess_text(review)
        review = re.sub('[^a-zA-Z]', ' ', review)
        review = re.sub('\s+',' ', review)
        review = review.lower()
        for word in review.split():
            if not word in stop_words:
                word = stemmer.stem(word)
                string += word + " "
        return string 
    
```

```python
train_data_cleaned['stopwords_remove_review']=train_data_cleaned['remove_punctuation_review'].apply(remove_stopwords)
```

```python
train_data_cleaned.head()
```

```python
from nltk.tokenize import word_tokenize
```

```python
def tokenize_word(text):
    return word_tokenize(text)
```

```python
train_data_cleaned['tokenized_review']=train_data_cleaned['stopwords_remove_review'].apply(tokenize_word)
```
